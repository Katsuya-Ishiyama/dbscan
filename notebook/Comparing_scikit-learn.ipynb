## DBSCAN using sklearn.cluster.DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
src_data = pd.read_csv('../data/sample_dataset.csv')
```

```python
X = src_data.loc[:, ['x', 'y']].values
```

```python
db = DBSCAN(eps=0.3, min_samples=10).fit(X)
```

```python
cluster_id = db.labels_
result_by_sklearn = src_data.rename(columns={'label': 'true_cluster_id'}).assign(cluster_id=cluster_id)
```

```python
result_by_sklearn.head()
```

```python
sns.scatterplot(x='x', y='y', hue='cluster_id', data=result_by_sklearn)
```
